# The baseline will recommend the most recent news articles that appeared in the dataset.

```python
import pandas as pd
import numpy as np

#Import the dataset
df = pd.read_csv('../dataset/corpus_with_parties_NE_24.10.2022.csv')
df = df.iloc[: , 2:]
df.head()
```

 Converting string data type into datetime object.

```python
# checking datatype
print(type(df.date[0]))
 
# convert to date
df['date'] = pd.to_datetime(df['date'])
 
# verify datatype
print(type(df.date[0]))

df=df.sort_values(by='date', ascending=False)

df[['title','parties']].head(10)
```

```python
df.head(10).to_csv('output/baseline_all_columns.csv')
```
