```python
import pandas as pd
import numpy as np
import ast

df = pd.read_csv('../../first_prototype/International_v/dataset/corpus_parties_NE_polarity_26.10.csv', index_col='id')

words_in_title=[]
char_in_title=[]
words_in_text=[]
char_in_text=[] 
words_in_author_bio=[]
char_in_author_bio=[]
num_of_NE=[]
num_of_authors=[]
polarity=[]

Date=[]


countOfWords=0
```

```python
features = ['title', 'text', 'author_bio', 'polarity', 'author']
for feature in features:
    df[feature]=df[feature].astype('str')
```

- Get number of words and characters in title, text, and author_bio
- Get number of NE andauthors, 

```python
for index, row in df.iterrows():
    
    #words in title
    countOfWords = len(row['title'].split()) if row['title'] !="nan" else 0
    words_in_title.append(countOfWords)
    
    #char in title
    countOfWords = len(row['title']) if row['title'] !="nan" else 0
    char_in_title.append(countOfWords)
    
    #words in text
    countOfWords = len(row['text'].split()) if row['text'] !="nan" else 0
    words_in_text.append(countOfWords)
    
    #char in text
    countOfWords = len(row['text']) if row['text'] !="nan" else 0
    char_in_text.append(countOfWords)
    
    #words in bio
    countOfWords = len(row['author_bio'].split()) if row['author_bio'] !="nan" else 0
    words_in_author_bio.append(countOfWords)
    
    #char in bio
    countOfWords = len(row['author_bio']) if row['author_bio'] !="nan" else 0
    char_in_author_bio.append(countOfWords)
    
    #num of NE
    countOfWords=len(row['NE'].split(","))
    num_of_NE.append(countOfWords)
    
    #num of authors
    countOfWords=0 if row['author'] =="nan" else len(row['author'].split(";"))
    num_of_authors.append(countOfWords)
    
    #polarity
    x=ast.literal_eval(row['polarity'])
    countOfWords=next(iter(x.values()))
    polarity.append(countOfWords)
```

The statistics

```python
ndf = pd.DataFrame(columns=['words_title', 'char_title', 'words_text', 'char_text', 'words_bio', 'char_bio', 'polarity', 'num_authors', 'NE']
)

ndf['words_title']=words_in_title
ndf['char_title']=char_in_title
ndf['words_text']=words_in_text
ndf['char_text']=char_in_text
ndf['words_bio']=words_in_author_bio
ndf['char_bio']=char_in_author_bio
ndf['NE']=num_of_NE
ndf['num_authors']=num_of_authors
ndf['polarity']=polarity
ndf.head(3)
```
