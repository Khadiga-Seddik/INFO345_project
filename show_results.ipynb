```python
import pandas as pd
pd.set_option('display.max_colwidth', 0)
```

```python
df_pol= pd.read_csv('../../first_prototype/international_v/dataset/corpus_parties_NE_polarity.csv')
df = pd.read_csv('output/baseline_all_columns.csv')
```

```python
df_pol=df_pol.set_index('id')
df.head(1)
df['id']= df['id'].astype('str')
polarity = []
df=df.reset_index(drop=True)
for index, row in df.iterrows():
    _id=row['id']
    doc = df_pol.loc[_id]
    polarity.append(doc['polarity'])

df['polarity'] = polarity
```
